## Q1

* As we have to do binary classification, we make use to Logistic Regression. Logistic Regression estimates the probability of a point belonging to a particular class. We apply sigmoid to the estimated values $w^TX$ to ensure that the values are in between 0 and 1. We apply cross entropy loss between the estimated values and true labels(either 0 or 1).Parameters are updated based on the gradients in order to minimize the Loss. 

## Q2
* The Loss function used is Cross Entropy Loss. It is given by

     $y_{true}(log(y_{pred}))  - (1-y_{true})(log(1-y_{pred}))$ 
* This is used as it is more efficient for binary classification

## Q3
* We can also used MSE Loss. But by using MSE there is chance of model to converge to its local minimum instead of global minimum. By using cross entropy Loss the model converges to global minimum.
